```python
from sklearn.datasets import load_iris
iris_data = load_iris()
df_iris = pd.DataFrame(data=iris_data.data, columns=iris_data.feature_names)
df_iris
```

```python
iris_data
```

```python
llaves = list(iris_data.keys())  #--> castear
```

```python
# Ciclo para recorrer:
targets = []
for t in iris_data.target:
  targets.append(t) #meter cosas a targets

targets
```

```python
targets = [iris_data.target_names[target] for target in iris_data.target]
targets
```

```python
df_iris.columns
```

```python
# Agregar una nueva columna a mi dataframe
df_iris["species"] = targets
df_iris.head(3)
```
